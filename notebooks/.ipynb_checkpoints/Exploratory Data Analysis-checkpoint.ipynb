```python
pristine_image_shapes = []
fake_image_shapes = []
mask_shapes = []


pristine_image_shapes_c = []
fake_image_shapes_c = []
mask_shapes_c = []


for i in os.listdir('../input/data/dataset-dist/phase-01/training/pristine/'):
    img = cv2.imread('../input/data/dataset-dist/phase-01/training/pristine/' + i)
    try:
        pristine_image_shapes.append(img.shape)
    except: 
        continue

for i in os.listdir('../input/groundtruths/first-image-forensics-masks/'):
    img = cv2.imread('../input/groundtruths/first-image-forensics-masks/' + i)
    try:
        mask_shapes.append(img.shape)
    except:
        continue

for i in os.listdir('../input/data/dataset-dist/phase-01/training/fake/'):
    img = cv2.imread('../input/data/dataset-dist/phase-01/training/fake/' + i)
    try:
        fake_image_shapes.append(img.shape)
    except:
        continue

for i in os.listdir('../input/data/Tp/'):
    img = cv2.imread('../input/data/Tp/' + i)
    try:
        pristine_image_shapes_c.append(img.shape)
    except:
        continue

for i in os.listdir('../input/groundtruths/casia-masks/'):
    img = cv2.imread('../input/groundtruths/casia-masks/' + i)
    try:
        mask_shapes_c.append(img.shape)
    except:
        continue

for i in os.listdir('../input/data/Au/'):
    img = cv2.imread('../input/data/Au/' + i)
    try:
        fake_image_shapes_c.append(img.shape)
    except:
        continue
```

```python
extns = []

for root, dirs, files in os.walk('../input/data/dataset-dist/phase-01/training/fake/'):
    for file in files:
        splitter = file.split('.')
        if len(splitter) > 0:
            extension = splitter[-1]
            extns.append(extension)

for root, dirs, files in os.walk('../input/data/dataset-dist/phase-01/training/pristine/'):
  for file in files:
    splitter = file.split('.')
    if len(splitter) > 0:
      extension = splitter[-1]
      extns.append(extension)
      
# for pristine images
for root, dirs, files in os.walk('../input/data/Au/'):
  for file in files:
    splitter = file.split('.')
    if len(splitter) > 0:
      extension = splitter[-1]
      extns.append(extension)

# for fake images
for root, dirs, files in os.walk('../input/data/Tp/'):
  for file in files:
    splitter = file.split('.')
    if len(splitter) > 0:
      extension = splitter[-1]
      extns.append(extension)
```

#### Random Model

```python
# creating csv file for given images
filenames = []
labels = []

# label all pristine images
for file in os.listdir('../input/data/dataset-dist/phase-01/training/pristine/'):
    filenames.append(file)
    labels.append(1)

# label all fake images
for file in os.listdir('../input/data/dataset-dist/phase-01/training/fake/'):
    filenames.append(file)
    labels.append(0)
    
# label all pristine images
for file in os.listdir('../input/data/Tp/'):
    filenames.append(file)
    labels.append(1)

# label all fake images
for file in os.listdir('../input/data/Au/'):
    filenames.append(file)
    labels.append(0)

# create a Dataframe and save it to csv file
df = pd.DataFrame(data={'files': filenames, 'labels': labels})
df.to_csv('./labels.csv', sep=',', index=False)
```

```python
df = pd.read_csv('./labels.csv')
```

```python
random_predictions = []
for file in df['files'].to_list():
    random_predictions.append(random.randint(0,1))
print('random predictions completed')
```

```python
metrics = MetricUtils()
print("log loss of random predictions: ", metrics.getLogLoss(df['labels'].to_list(), random_predictions))
```

```python
fpr, tpr, thresholds = metrics.getROCValues(df['labels'].to_list(), random_predictions)
```

```python
plt.plot(fpr, tpr)
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.title('ROC curve')
plt.show()
```
